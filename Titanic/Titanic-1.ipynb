```python
import pandas as pd
import numpy as np
```

```python
tit_data=pd.read_csv("train.csv")
```

```python
tit_data.head()
```

```python
tit_data.info()
```

```python
tit_data.drop(['Embarked'],axis=1,inplace=True)
```

```python
tit_data.drop(['Cabin'],axis=1,inplace=True)
```

```python
tit_data.drop(['Ticket'],axis=1,inplace=True)
```

```python
tit_data.drop(['Name'],axis=1,inplace=True)
```

```python
tit_data['Sex'] = tit_data['Sex'].map({'male': 1, 'female': 0})
```

```python
tit_data.isnull().sum()
```

```python
tit_data=tit_data[~np.isnan(tit_data['Age'])]
```

```python
St_titdata=tit_data[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare']]
```

```python
St_titdata=(St_titdata-St_titdata.mean())/St_titdata.std()
```

```python
St_titdata.head()
```

```python
Sur_titdata=tit_data['Survived']
```

```python
Sur_titdata.head()
```

```python
import statsmodels.api as sm
```

```python
logm1 = sm.GLM(Sur_titdata,(sm.add_constant(St_titdata)), family = sm.families.Binomial())
logm1.fit().summary()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
from sklearn.feature_selection import RFE
rfe = RFE(logreg, 4)             # running RFE with 13 variables as output
rfe = rfe.fit(St_titdata,Sur_titdata)
print(rfe.support_)           # Printing the boolean results
print(rfe.ranking_) 
```

```python
St_titdata.columns[rfe.support_]
```
